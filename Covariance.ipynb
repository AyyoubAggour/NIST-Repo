```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import LogNorm
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('L0001.csv')
```

```python
cov = df.cov()


plt.figure(figsize=(10, 8))


sns.heatmap(cov, annot=True, cmap='viridis', norm=LogNorm())


plt.title('Covariance Matrix Heatmap', fontsize=12)

x = np.linspace(0, 100, 100)  # Generate 100 points between 0 and 10
y = x # Define a linear relationship
plt.plot(x, y, 'k--')

plt.show() #WHITE VALUES DISPLAY NEGATIVES
```

```python
cov = df.cov()


plt.figure(figsize=(10, 8))

df_negative_only = cov.copy()
df_negative_only[df_negative_only > 0] = np.nan


sns.heatmap(df_negative_only, annot=True, cmap='viridis', center=0, cbar=True)


x = np.linspace(0, 100, 100)  # Generate 100 points between 0 and 10
y = x # Define a linear relationship
plt.plot(x, y, 'k--')

plt.title('Covariance Matrix Heatmap', fontsize=12)

plt.show() #WHITE VALUES DISPLAY POSITIVES
```
