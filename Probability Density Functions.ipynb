```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
df = pd.read_csv('L0001.csv')
```

```python
mu = -1  # mean
sigma = 1  # standard deviation

x = np.linspace(-5, 5, 1000)  # x-values for the plot

pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))
```

```python
plt.figure(figsize=(8, 6))  # Adjust the figure size (optional)

plt.plot(x, pdf, label='Normal Distribution PDF')
plt.title('Probability Density Function of a Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def graphProb(par1):

    var = df[par1]
    
    mu = var.mean().round(3)  # mean
    print(f'Mean: {mu}')
    sigma = var.std()  # standard deviation
    print(f'Standard Deviation: {sigma}')

    x = np.linspace(var.min(), var.max(), 10000)  # x-values for the plot

    pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))
    #print(pdf)

    
    plt.figure(figsize=(8, 6))  # Adjust the figure size (optional)

    plt.hist(var, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
    plt.plot(x, pdf, label='Normal Distribution PDF', c='red')
    plt.title('Probability Density Function of a Normal Distribution')
    plt.xlabel(par1)
    #plt.ylim(0,0.001)
    #plt.xlim()
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
```
