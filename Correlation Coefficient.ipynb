```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import math
from sklearn import metrics
from scipy.stats import pearsonr, t

#Calculates correlation coefficient
#closer r is to 1 the more they relate if its closer to 0 they dont really relate
```

```python
df = pd.read_csv('L0001.csv') #you can also enter whatever dataset you want
df.fillna(0, inplace=True) #all NaN values become 0's
```

```python
def cc(par1, par2, alpha):

    n = len(df[par1])

    x = df[par1].sum()
    y = df[par2].sum()

    #print(f'Sum of Variable 1: {x}')
    #print(f'Sum of Variable 2: {y}')

    xy = ((df[par1]*df[par2]).sum())
    
    #print(f'Sum of Variable 1 * Variable 2: {xy}')

    x2 = ((df[par1]*df[par1]).sum())
    #print(f'Sum of Variable 1 Squared: {x2}')

    y2 = ((df[par2]*df[par2]).sum())
    #print(f'Sum of Variable 2 Squared: {y2}')

    #print(f'Number of data points: {n}')

    r = (((n*xy)-(x*y))/math.sqrt(((n*x2)-x2)*((n*y2)-y2)))
    #print(f'R Value: {r}')

    #Critical Value Calculation


    #t = (r*(math.sqrt(n-2))/math.sqrt(1-(r*r)))
    #print(t)

    
    freedoms = n - 2 #degrees of freedom
  
    critv = t.ppf(1 - alpha/2, freedoms)

    print(f'Critical Value: {critv}')

    if abs(r) > abs(critv):
        print(f'Correlation between {par1} and {par2}:')
        print(f'R value: {r}')
        print(f'Critical Value: {critv}')
```

```python
cc('x','y', 0.5)
```

```python
for i in range(len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        col1 = df.columns[i]
        col2 = df.columns[j]
        cc(col1, col2, 0.5)
       
```

```python
#COMPLETE FUNCTION COMPARING ALL COLUMNS WITH EACH OTHER CHECKING IF CORRELATION COEFFICIENT (R) IS GREATER THAN CRITICAL VALUE
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import math
from sklearn import metrics
from scipy.stats import pearsonr, t

df = pd.read_csv('L0001.csv')

df.fillna(0, inplace=True) #all NaN values become 0's

#Calculate critical value
def critvalue(par1, par2, alpha):
    
    n = len(df[par1])

    x = df[par1].sum()
    y = df[par2].sum()

    #print(f'Sum of Variable 1: {x}')
    #print(f'Sum of Variable 2: {y}')

    xy = ((df[par1]*df[par2]).sum())
    
    #print(f'Sum of Variable 1 * Variable 2: {xy}')

    x2 = ((df[par1]*df[par1]).sum())
    #print(f'Sum of Variable 1 Squared: {x2}')

    y2 = ((df[par2]*df[par2]).sum())
    #print(f'Sum of Variable 2 Squared: {y2}')

    #print(f'Number of data points: {n}')

    #my way 
    #r = (((n*xy)-(x*y))/math.sqrt(abs(((n*x2)-x2)*((n*y2)-y2))))

    #premade function way
    r, _ = pearsonr(df[par1], df[par2])
    
    freedoms = n - 2 #degrees of freedom
  
    critv = t.ppf(1 - alpha/2, freedoms)

    #print(f'Critical Value: {critv}')

    if abs(r) > abs(critv):
        print(f'Correlation between {par1} and {par2}:')
        print(f'R value: {abs(r)}')
        print(f'Critical Value: {critv}')



for i in range(len(df.columns)):
    for j in range(len(df.columns)):
        if i != j:  # Ensure we don't compare a column with itself
            col1 = df.columns[i]
            col2 = df.columns[j]
            critvalue(col1, col2, 0.5)
```
