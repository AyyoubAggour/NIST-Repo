```python
import numpy as np

np.random.seed(0)

df = pd.DataFrame({'x' : np.random.randn(500), 'MPArea' : np.random.randn(500)})

X = df['x']
y = df['MPArea']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 203)

from sklearn.linear_model import LinearRegression

X_train = np.array(X_train).reshape(-1,1)
X_train


```

```python
X_test = np.array(X_test).reshape(-1,1)
X_test.flatten()
```

```python
clf = LinearRegression()

clf.fit(X_train, y_train)
```

```python
c = clf.intercept_
c #y-int
```

```python
m = clf.coef_
m #slope
```

```python
y_pred_train = m*X_train + c
y_pred_train.flatten()

```

```python
clf.predict(X_train)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred_train, color='red')
plt.grid(True)

plt.xlabel('x')
plt.ylabel('MPArea')
plt.title('X vs Melt Pool Area')
plt.xticks(np.arange(-3, 4, 0.5))
plt.yticks(np.arange(-4,4,0.5))
plt.xticks(rotation=25) #rotate x ticks 25 degrees from the horizontal
plt.show()


#END PRODUCT - Linear Regression using 500 random data points comparing X and Melt Pool Area
```
