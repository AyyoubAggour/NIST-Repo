```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.covariance import LedoitWolf
from scipy.linalg import eigh
from sklearn.decomposition import PCA

def pca_covariance_error(X, n_components=None):
    
    
    scaler = StandardScaler()
    X_standardized = scaler.fit_transform(X)
    

    X_centered = X_standardized - np.mean(X_standardized, axis=0)
    

    covariance_matrix = LedoitWolf().fit(X_centered).covariance_
    

    eigenvalues, eigenvectors = eigh(covariance_matrix)
    

    eigenvalues = np.maximum(eigenvalues, 0)
    

    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]
    

    if n_components is not None:
        eigenvalues = eigenvalues[:n_components]
        eigenvectors = eigenvectors[:, :n_components]
    

    explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
    
    X_reduced = X_centered @ eigenvectors
    X_reconstructed = X_reduced @ eigenvectors.T
    reconstruction_error = np.mean(np.square(X_centered - X_reconstructed))
    

    if np.isnan(reconstruction_error):
        raise ValueError("Reconstruction error resulted in NaN. This may indicate numerical instability or an issue with the data.")
    
 

#Reconstruction Error - The reconstruction error measures how well the original data can be reconstructed using a reduced number of principal components.
#Low Error: A low reconstruction error indicates that the reduced components capture most of the important information in the data.
#High Error: A high error suggests that the reduced components may not adequately capture the variance,--
# --meaning you might need more components for accurate reconstruction.

#Explained Variance Ratio - The explained variance ratio for each principal component shows the proportion--
# --of the total variance in the data that is captured by that component.
#A component with an explained variance ratio of 0.9 or higher (when considering the cumulative explained variance)--
# --is often considered to capture a substantial amount of the variance.

    
    # Perform PCA
    '''
    pca = PCA()
    pca.fit(X)

    # Get eigenvalues (explained variance)
    explained_variances = pca.explained_variance_

    # Compute explained variance ratio
    explained_variance_ratios = pca.explained_variance_ratio_

    # Print results
    print("Explained Variances:", explained_variances)
    print("Explained Variance Ratios:", explained_variance_ratios)

    # Calculate cumulative explained variance ratio
    cumulative_variance_ratio = np.cumsum(explained_variance_ratios)
    print("Cumulative Explained Variance Ratio:", cumulative_variance_ratio)
    '''

    return reconstruction_error, explained_variance_ratio


X = df
reconstruction_error, explained_variance_ratio = pca_covariance_error(X, n_components=5)
    
print("Reconstruction Error:", reconstruction_error)
print("Explained Variance Ratio:", explained_variance_ratio)

```
