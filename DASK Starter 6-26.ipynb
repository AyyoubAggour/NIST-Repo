```python
#import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()


df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

print(df.head())

import matplotlib.pyplot as plt

# Scatter plot of sepal length vs sepal width
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=iris.target, cmap='viridis')
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Species', ticks=range(3), orientation='horizontal', extend='both')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


iris = load_iris()


df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

mean_values = df.groupby('species').mean()

# Plotting the line graph
plt.figure(figsize=(10, 2))

# Plot each feature's mean values across species
for feature in iris.feature_names:
    plt.plot(mean_values.index, mean_values[feature], marker='o', label=feature)

plt.title('Mean Feature Values across Iris Species')
plt.xlabel('Species')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=5)
plt.tight_layout()

plt.show()

```
