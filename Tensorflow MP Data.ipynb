```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('L0001.csv')


X = pd.get_dummies(df.drop(['MPArea', 'x'], axis = 1))
y = df['MPArea'].apply(lambda x: 1 if x > 360  else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns))) #units = # of neurons in NN, 
model.add(Dense(units=64, activation='relu')) #neural network (NN) hidden layers
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, batch_size = 32) #epoch = time trained
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```
